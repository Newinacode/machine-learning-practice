```python
# to create more sophisticated plots we can use contour() method that produces a contour plot in order to represent 3D (three dimensional
# data, similar to a topographical map
# It takes three arguments
# 1. A vector of x values (first dimension),
# 2. A vector of y values (second dimension),
# 3. A matrix whoes elements correspond to the z values -> Third dimension -> for each pari of (x,y) coordinates
```

```python
fig,ax = subplots(figsize=(8,8))
# linspace(a,b,n) create a vector from a to b with n numbers
x = np.linspace(-np.pi,np.pi,50)
x
y = x
f = np.multiply.outer(np.cos(y),1/(1+x**2))
fig,ax = subplots(figsize=(8,8))
ax.contour(x,y,f);
```

```python
fig,ax  = subplots(figsize=(8,8))
ax.contour(x,y,f,levels=10);
```

```python
#ax.imshow() method is similar to ax.contour(), except that it produces a color-coded plot whose colors depend on the z value. This is known
#  as heatmap and it sometimes used to plot temperature in weather forecasts
fig, ax = subplots(figsize=(8,8))
ax.imshow(f);
```

```python
# Indexing data
A = np.array(np.arange(16)).reshape(4,4)
```

```python
A
```

```python
A[2,2] #first will index row and second will index columns
```

```python
# Indexing rows, columns and submatrices
```

```python
A[[1,3]]
```

```python
#  this will return submatrices row -> 1 and 3
```

```python
# this will include first and second columns
A[:,[0,2]]
```

```python
# if we want to include only 2nd and 3rd rows and only two columns i.e. 1st and second
A[[1,2],[0,2]]
```

```python
# we except to get submatrices but only get one dimensional output, why when we pass two value it is treated as x,y so we need to create
# one sub-matrix for row and apply to it example
A[[1,2]][:,[0,2]]
```

```python
# we can get same thing with help of convenience function np.ix_
```

```python
idx = np.ix_([1,2],[0,2])
A[idx]
```

```python
# or we can do that by indexing
A[1:3:1,0:3:2]  #[x:y:s,p:q:s] here,x,y is start and end for row, and p and q is start and end for column where s in how many steps.
# so here 1,3 first and second is include
# for cols 0 two step then col 2 is included
```

```python
# boolean indexing
keep_rows = np.zeros(A.shape[0],bool)
```

```python
keep_rows
```

```python
# This is going to step first and 3rd element as true
keep_rows[[1,3]] = True
keep_rows[0] = True 
keep_rows
```
