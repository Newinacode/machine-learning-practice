```python
rng = np.random.default_rng(1)
A = rng.standard_normal((127,5))
```

```python
M = rng.choice([0,np.nan],p=[0.8,0.2],size=A.shape)
```

```python
M
```

```python
A += M
```

```python
D = pd.DataFrame(A,columns=['food',
'bar',
'pickle',
'snack',
'popcorn'])
```

```python
D
```

```python
D[:3]
```

```python
for col in D.columns:
    template = "{0} is missing {1:.2%} data"
    mean = np.isnan(D[col]).mean()
    print(template.format(col,mean))
```
